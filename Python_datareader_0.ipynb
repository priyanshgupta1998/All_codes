#Multiple company's Stock price Analysis   
simultaneously we can see the live status of the stock market of more than one companies. and can  distinguish between them.   
As of now , we are taking 4 compies :  
  facebook , amazon , netflix , google 

````
stocks = ['FB' , 'AMZN' , 'NFLX' , 'GOOG']   
````

````
data = pdr.get_data_yahoo(stocks , start = '2019-02-21')  # from 21 february 2019
print(data.shape)
print(datetime.datetime.now())
data   # Can see live status of stock market & prices   
````

````
data['High'].tail()   # High Stock prices of each companies in current days.
````
