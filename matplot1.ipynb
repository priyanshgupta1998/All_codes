```
import matplotlib.pyplot as plt
```

```
plt.plot([1,2,3,4,5])  # now we are passing a list of y axis variables 
```

```
plt.plot([1,2,3,4,5],'ro-')
```

```
plt.plot([1,2,3,4,5],[1,4,9,16,25])  # X,Y 
```

```
plt.plot([1,2,3,4,5],[1,4,9,16,25],'go-')
```

```
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ys-')  # square 
```

```
plt.plot([1,2,3,4,5],[1,4,9,16,25],'b^')   # triangle
plt.grid(False)
```

```
plt.xlabel('the values ' , fontsize = 18 , color = 'green')
plt.ylabel('the squares ' , fontsize = 18 , color = 'red')
plt.title( 'Graph')
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ys-')
```

```
plt.xlabel('the values ' , fontsize = 18 , color = 'green')
plt.ylabel('the squares ' , fontsize = 18 , color = 'red')
plt.title( 'Graph')
plt.axis([0,6,0,30])
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ys-')
```

```
plt.xlabel('the values ' , fontsize = 18 , color = 'green')
plt.ylabel('the squares ' , fontsize = 18 , color = 'red')
plt.title( 'Graph')
plt.axis([0,6,0,30])
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ys-' , linewidth= 3.0)  # by default it is 1.0
```

```
plt.xlabel('the values ' , fontsize = 18 , color = 'green')
plt.ylabel('the squares ' , fontsize = 18 , color = 'red')
plt.title( 'Graph')
plt.grid(True)  # By default it is False(off)
plt.axis([0,6,0,30])
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ys-' , linewidth= 3.0)  # by default it is 1.0
```

# Scatter plot

```
import random
x = [random.randint(1,100) for i in range(100)]
y = [random.randint(1,100) for i in range(100)]
plt.axis([0,500,0,500])
plt.scatter(x,y,color='red')
```
