#Convert into BLUE (BGR)

````
image2 = np.array([blue_data, green_data , red_data ])
image2 = np.swapaxes(image2 , 0 ,1)
image2 = np.swapaxes(image2  , 1,2)
cmap = 'hot'
plt.figure(figsize = (15,10))
plt.subplot(232)
plt.imshow(image2  , cmap = cmap , interpolation = 'nearest')
plt.axis('off')
````

#Light Green ( BRG)

````
image3 = np.array([blue_data, red_data, green_data  ])
image3  = np.swapaxes(image3  , 0 ,1)
image3  = np.swapaxes(image3  , 1,2)
cmap = 'hot'
plt.figure(figsize = (15,10))
plt.subplot(232)
plt.imshow(image3  , cmap = cmap , interpolation = 'nearest')
plt.axis('off')
plt.title('Original image in RGB')
````

# RBG

````
image4 = np.array([ red_data, blue_data,green_data  ])
image4= np.swapaxes(image4 , 0 ,1)
image4 = np.swapaxes(image4 , 1,2)
cmap = 'hot'
plt.figure(figsize = (15,10))
plt.subplot(232)
plt.imshow(image4 , cmap = cmap , interpolation = 'nearest')
plt.axis('off')
plt.title('Original image in RGB')
````

# Different type of coloured image

````
image5 = np.array([red_data , 256*green_data, 256*blue_data  ])
image5 = np.swapaxes(image5  , 0 ,1)
image5  = np.swapaxes(image5  , 1,2)
cmap = 'hot'
plt.figure(figsize = (15,10))
plt.subplot(232)
plt.imshow(image5 , cmap = cmap , interpolation = 'nearest')
plt.axis('off')
plt.title('Original image in RGB')
````

#image with specific background

````
#image6 = np.array([ 1.0*green_data, 0.5*red_data , 0.5*blue_data  ])
image6 = np.array([(0.5*green_data)%256 ,(0.5*red_data)%256,(0.5*blue_data)%256 ]).astype(np.uint8)
image6 = np.swapaxes(image6 , 0 ,1)
image6  = np.swapaxes(image6  , 1,2)
cmap = 'hot'
plt.figure(figsize = (15,10))
plt.subplot(232)
plt.imshow(image6 , cmap = cmap , interpolation = 'nearest')
plt.axis('off')
plt.title('Original image in RGB')
````

````
image6 = np.array([(1.5*green_data)%256 ,(1.0*red_data)%256,(1.0*blue_data)%256 ]).astype(np.uint8)
image6 = np.swapaxes(image6 , 0 ,1)
image6  = np.swapaxes(image6  , 1,2)
cmap = 'hot'
plt.figure(figsize = (15,10))
plt.subplot(232)
plt.imshow(image6 , cmap = cmap , interpolation = 'nearest')
plt.axis('off')
plt.title('Original image in RGB')
````
